```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = 'matches.csv'
ipldf = pd.read_csv(data)
```

```python
ipldf.head()
ipldf.info()
```

```python
team_wins = ipldf['winner'].value_counts()
print(team_wins)
```

```python
player_of_match_counts = ipldf['player_of_match'].value_counts()
print(player_of_match_counts)
```

```python
average_runs_per_match = ipldf['target_runs'].mean()
print(average_runs_per_match)
```

```python
plt.figure(figsize= (10, 6))
sns.countplot(x= 'winner', data = ipldf)
plt.title("Number of wins by Teams")
plt.xlabel("Team")
plt.ylabel("Number of wins")
plt.show()
```

```python
plt.figure(figsize= (8, 5))
plt.hist(ipldf['target_runs'], bins = 10)
plt.title("Distibution of Runs Scored per Match")
plt.xlabel("Runs Scored")
plt.ylabel("Frequency")
plt.show()
```
